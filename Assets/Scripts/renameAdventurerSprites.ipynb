```python
import os
import pandas as pd

# Define the path to the folder
folder_path = r"C:\Users\Benjamin\Desktop\Unity\TheArcher\Assets\Art\Final\Player\Adventurer-1.5\Edited Individual Sprites"

# Define the list of animations and their ranges
animations = {
    "idle": (1, 4),
    "crouch": (5, 9),
    "run": (10, 15),
    "jump": (16, 19),
    "smrslt": (20, 23),
    "fall": (24, 25),
    "slide": (26, 27),
    "stand": (28, 30),
    "crnr-grb": (31, 34),
    "crnr-clmb": (35, 39),
    "idle-2": (40, 43),
    "attack1": (44, 48),
    "attack2": (49, 54),
    "attack3": (55, 60),
    "hurt": (61, 63),
    "die": (64, 70),
    "swd-drw": (71, 74),
    "swd-shte": (75, 78),
    "crnr-jmp": (79, 80),
    "wall-slide": (81, 82),
    "ladder-climb": (83, 86),
    "cast": (87, 90),
    "cast_loop": (91, 94),
    "items": (95, 97),
    "air-attack1": (98, 101),
    "air-attack2": (102, 104),
    "air-attck3-rdy": (105, 105),
    "air-attack3_loop": (106, 107),
    "air-attack-3-end": (108, 110),
}

# Create a list to hold the old and new file names for visualization
rename_data = []

# Rename files in the folder
for filename in os.listdir(folder_path):
    if filename.startswith("adventurer-v1_") and filename.endswith(".png"):
        # Extract the number from the file name
        number = int(filename.split('_')[1].split('.')[0])
        
        # Find the appropriate animation name and index
        for animation, (start, end) in animations.items():
            if start <= number <= end:
                # Calculate the new index for the animation
                new_index = number - start
                
                # Build the new file name
                new_filename = f"adventurer-{animation}-{new_index:02}.png"
                
                # Rename the file
                old_path = os.path.join(folder_path, filename)
                new_path = os.path.join(folder_path, new_filename)
                os.rename(old_path, new_path)
                
                # Add the renaming info to the list for display
                rename_data.append({"Old Filename": filename, "New Filename": new_filename})
                break

# Create a DataFrame to display the renaming results
rename_df = pd.DataFrame(rename_data)

# Display the DataFrame
import ace_tools as tools; tools.display_dataframe_to_user(name="Renamed Files", dataframe=rename_df)
```
